## Simple Data Manipulation

```python
data.describe()
```

```python
data['hour_lambda'] = data['Time (UTC)'].apply(lambda x: x.hour)
```

```python
def hrs(x):
    return x.dt.hour
```

```python
data['hours_func'] = hrs(data['Time (UTC)'])
```

```python
data['hours_assign'] = data['Time (UTC)'].dt.hour
```

```python
data.describe()
```

```python
data.groupby('hour_lambda').describe()
```
